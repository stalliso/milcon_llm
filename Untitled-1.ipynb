```python
import pdfplumber
import pandas as pd
import re

def pdf_toc_to_dataframe(pdf_path):
    """
    Extract Table of Contents from a PDF into a DataFrame.
    Works for both embedded TOC (bookmarks) and text-based TOC pages.
    """
    
    # ── Method 1: Embedded bookmarks (works if PDF has bookmarks) ──────────────
    import fitz  # PyMuPDF
    
    def get_bookmarks(pdf_path):
        doc = fitz.open(pdf_path)
        toc = doc.get_toc()  # returns [[level, title, page], ...]
        doc.close()
        if toc:
            df = pd.DataFrame(toc, columns=["Level", "Title", "Page"])
            return df
        return None

    # ── Method 2: Parse text-based TOC page ────────────────────────────────────
    def get_text_toc(pdf_path):
        rows = []
        with pdfplumber.open(pdf_path) as pdf:
            for page in pdf.pages[:10]:  # TOC usually in first 10 pages
                text = page.extract_text()
                if not text:
                    continue
                lines = text.split("\n")
                for line in lines:
                    # Match patterns like: "1. Introduction ........... 5"
                    # or "Chapter 1  Introduction  5"
                    match = re.match(
                        r"^(\d+[\.\d]*\s+)?(.+?)\s+[.\s]{2,}\s*(\d+)\s*$", line.strip()
                    )
                    if match:
                        rows.append({
                            "Section": match.group(1).strip() if match.group(1) else "",
                            "Title":   match.group(2).strip(),
                            "Page":    int(match.group(3))
                        })
        return pd.DataFrame(rows) if rows else None

    # Try bookmarks first, fall back to text parsing
    df = get_bookmarks(pdf_path)
    if df is not None and not df.empty:
        print("✅ Extracted TOC from embedded bookmarks")
        return df

    df = get_text_toc(pdf_path)
    if df is not None and not df.empty:
        print("✅ Extracted TOC from text parsing")
        return df

    print("⚠️  No TOC found. Try adjusting the regex for your PDF's format.")
    return pd.DataFrame()


```

```python
# ── Usage ──────────────────────────────────────────────────────────────────────
df = pdf_toc_to_dataframe("your_file.pdf")
print(df)
df.to_csv("toc_output.csv", index=False)
```
