```python
# helper_code/export_eval_from_csv.py

import csv
import sys
from pathlib import Path
from collections import defaultdict

sys.path.append(str(Path.cwd().parent)) 

from helper_code.export_eval import export_eval_summary_png
from eval.eval_multi_part import MULTI_PART_QUESTIONS, EXPECTED_ROUTES

# ---- SET YOUR FILE PATHS HERE ----
QA_RESULTS_CSV  = "qa_results_20260313_131711.csv"
QA_MISSED_CSV   = "qa_missed_20260313_131711.csv"
ROUTING_CSV     = "routing_results_20260313_131711.csv"
OUTPUT_PNG      = "eval_summary_regenerated.png"
LAYOUT          = "wide"   # "standard" or "wide"
# -----------------------------------

with open(QA_RESULTS_CSV) as f:
    qa_accuracies = [float(row["accuracy"]) for row in csv.DictReader(f)]

qa_miss_counts = defaultdict(int)
with open(QA_MISSED_CSV) as f:
    rows = list(csv.DictReader(f))
    num_qa_runs = int(rows[0]["runs"])
    for row in rows:
        qa_miss_counts[row["question"]] = int(row["miss_count"])

q_pass_counts = defaultdict(int)
with open(ROUTING_CSV) as f:
    rows = list(csv.DictReader(f))
    num_multi_runs = int(rows[0]["runs"])
    for row in rows:
        q_pass_counts[row["question"]] = int(row["passes"])

run_accuracies = [q_pass_counts.get(q, 0) / num_multi_runs
                  for q in sorted(MULTI_PART_QUESTIONS.keys())]

export_eval_summary_png(
    qa_accuracies=qa_accuracies,
    run_accuracies=run_accuracies,
    q_pass_counts=q_pass_counts,
    qa_miss_counts=qa_miss_counts,
    multi_part_questions=MULTI_PART_QUESTIONS,
    expected_routes=EXPECTED_ROUTES,
    num_multi_runs=num_multi_runs,
    num_qa_questions=10,
    output_path=OUTPUT_PNG,
    layout=LAYOUT,
)

print(f"PNG saved -> {OUTPUT_PNG}")
```
