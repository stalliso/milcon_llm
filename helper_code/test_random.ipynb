```python
import os
import random

directory = "../docs/projects"
projects = [file for file in os.listdir(directory) if file.endswith(".pdf") and 'toc' not in file.lower() and 'ipl' not in file.lower()]

mcon = [file for file in projects if file.startswith("P")]
fsrm = [file for file in projects if file not in mcon]

mcon_test = random.sample(mcon, int(len(mcon)*.2))
fsrm_test = random.sample(fsrm, int(len(fsrm)*.2))

test_dir = "../docs/test"
test_proj = [file for file in os.listdir(test_dir)]

if len(test_proj) > 0:
    print("Test directory is not empty. Please clear it before running this code.")

else:
    os.makedirs(test_dir, exist_ok=True)
    for file in mcon_test:
        os.rename(os.path.join(directory, file), os.path.join(test_dir, file))
    for file in fsrm_test:
        os.rename(os.path.join(directory, file), os.path.join(test_dir, file))
```
